```python
class Evolver():
    def __init__(self, pop_size, retain_proportion=0.2, random_select=0.05, mutate_prob=0.01):
        self.pop_size = pop_size
        self.retain = int(retain_proportion * pop_size)
        self.random_select = random_select
        self.mutate_prob = mutate_prob
        self.splice = 0
        self.eps = 1e-8
        self.inf = 9999999999999999.0
        self.fitness = lambda x,y: abs(y - sum([i**2 for i in x]))
        self.termination_score = 0.01
        
        
    def evolve(self, n, y):
        self.splice = n//2
        pop = self.__init_population(n, int(y**0.5))
        score = self.inf
        pop_score = self.inf
        i=0
        while pop_score > self.termination_score:
            i+=1
            if i>100000:
                break
                
            pop, pop_score, score = self.__evolve_population(pop, y)
            
            if i%1000==0:
                pop = sorted(pop, key=lambda x: x['f'])
                print(i, 'best', list(pop)[0], pop_score, score)
            
        results = set([', '.join([str(x) for x in ind['x']]) for ind in pop if ind['f']<=self.eps])
        
        print('Iter: %d, Score: %.2f, Results: %s' % (i, score, str(results)))
        
        return results
    
    
    def __init_population(self, n, rng):
        return [{'x':[random.randint(1,rng) for j in range(n)], 'f':self.inf} for i in range(self.pop_size)]
                
        
    def __evolve_population(self, pop, y):
        parents = self.__get_parents(pop)
        num_children = self.pop_size - len(parents)
        
        for _ in range(num_children):
            a,b = random.choice(parents),random.choice(parents)
            if a['x']==b['x']:
                b = random.choice(parents)
                
            child = {'x':a['x'][:self.splice] + b['x'][self.splice:], 'f':self.inf}
            
            if random.random() < self.mutate_prob:
                idx = random.randint(0, len(child['x'])-1)
                child['x'][idx] = random.randint(0, int(y**0.5))
        
            parents.append(child)
            
        parents = self.__calc_fitness(parents, y)
        
        score = min([p['f'] for p in parents])
        
        return parents, self.__score_population(parents), score
    
    
    def __score_population(self, pop):
        return sum([ind['f'] for ind in pop])/len(pop)  
    
    
    def __get_parents(self, pop):
        pop = sorted(pop, key=lambda x: x['f'])
        
        parents = []
        for p in list(pop):
            if not p in parents:
                parents.append(p)
                if len(parents)==self.retain:
                    break
        
        parents.extend([ind for ind in pop[self.retain:] if random.random() < self.random_select])
        
        return parents
    
        
    def __calc_fitness(self, pop, y):
        for ind in pop:
            ind['f'] = self.fitness(ind['x'], y)
        return pop
```
